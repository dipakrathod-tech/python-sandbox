# Attribute - state (data), class or object variable
- if the variable declared in body of class its class attibute/variale
- and if declared inside __init__() method of class using self that was instance(object) attibute/variable

```python
class Car:
    manfacturer = "tata"
    def __init__(self,model):
        self.model = model
    
        
    
```

```python
my_car = Car("example")
```

```python
print(f" class attributes value is : \"{my_car.manfacturer}\" and instance at attribute value: \"{my_car.model}\"")
```

### NOTE: __init__() takes first parameter as the reference of the object


```python
my_car2 = Car.__new__(Car)
Car.__init__(my_car2,"example2")
```

```python
print(my_car2.model)
```

- each object has its own copy of model That's the "state"

# Method ( Funtion )  - behavior

## Instance Method 
- An instance method is a function defined inside a class that, when accessed through an instance, automatically receives that instance as its first argument.
- defined as in method's first parameter using ( self by conventions )

```python
class Car:
    manfacturer = "tata"
    def __init__(self,model):
        self.model = model
    def start(self):
        print("engine is starting.....")
```

```python
new_car = Car("corolla")
```

```python
new_car.start()
```

## class method
- A class method is a method that is bound to the class, not the instance.
- It is defined using the @classmethod decorator.
- The first parameter (cls by convention) receives the class object automatically.

```python
class Car: 
    wheels = 4
    

    @classmethod
    def info(cls):
        return cls.wheels
        
```

```python
new_car = Car()
print(new_car.info())
```

## static methods

```python
class Car: 
    @staticmethod
    def info():
        return "just information"
        
```

```python
print(Car.info())
```

# Four Pillors of OOPs

## INHERITANCE 

```python
class Animal:
    def speak(self):
        print("some sound")

class Dog(Animal):
    def speak(self):
        super().speak()
        print("bark")
```

```python
my_dog = Dog()
my_dog.speak()
```

### Multiple Inheritance & MRO ( method resulation order )

```python
class A:
    pass
class B(A):
    pass
class C(B):
    pass
print(C.mro())
```

### Encapsulation

there is two thing important here: 
- safety and control


- PROTECTED: _ subclass/internal "by conventional not force"
- Class only __ do not touch (Strict)

```python
class Student: 
    def __init__(self, name):
        self.__project = "AI System"
    def value(self):
        print(self.__project)
        return self.__project
        
    
```
